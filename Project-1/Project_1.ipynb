# Преобразование данных

1. Признак **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
    * Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
    * Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…

Нас интересует только уровень образования. 

Создадим новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

```python
# создаем функцию
def get_education_type(education):
    education_list = education.split()
    education_type = education_list[0] + " " + education_list[1]
    if education_list[1] == 'образование':
        education_type = education_list[0]    
    return education_type

# создаем новый признак с применением функции
data['Образование'] = data['Образование и ВУЗ'].apply(get_education_type)

print(f"Количество соискателей, имеющих средний уровень образования: {data[data['Образование'] == 'Среднее'].shape[0]}")
```

2. Cтолбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
    * Мужчина , 39 лет , родился 27 ноября 1979 
    * Женщина , 21 год , родилась 13 января 2000

Создадим два новых признака **"Пол"** и **"Возраст"**. 

При этом важно учесть:

* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

```python
# создаем новые признаки
data['Пол'] = data['Пол, возраст'].apply(lambda x: 'М' if x.split()[0] == 'Мужчина' else 'Ж')
data['Возраст'] = data['Пол, возраст'].apply(lambda x: int(x.split()[2]))

print(f"Доля женских резюме: {round(data['Пол'].value_counts(normalize=True)['Ж'] * 100, 2)}%")
print(f"Средний возраст соискателей: {round(data['Возраст'].mean(), 2)} года")
```

3. Признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)".

При этом важно учесть:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

```python
# создаем функцию
def get_experience(arg):
    if arg is np.nan or arg == 'Не указано':
        return None
    month_key_words = ['месяц', 'месяцев', 'месяца']
    year_key_words = ['год', 'лет', 'года']
    arg_splitted = arg.split()[:7]
    years = 0
    months = 0
    for index, item in enumerate(arg_splitted):
        if item in year_key_words:
            years = int(arg_splitted[index-1])
        if item in month_key_words:
            months = int(arg_splitted[index-1])
    return int(years * 12 + months)

# создаем новый признак с применением функции
data['Опыт работы (месяц)'] = data['Опыт работы'].apply(get_experience)

print(f"Медианный опыт работы: {data['Опыт работы (месяц)'].median()} месяцев")
```

4. Признак "Город, переезд, командировки". Информация в нем представлена в следующем виде: **<Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>**. В скобках указаны необязательные параметры строки. Например, можно встретить следующие варианты:

* Москва , не готов к переезду , готов к командировкам
* Москва , м. Беломорская , не готов к переезду, не готов к командировкам
* Воронеж , готов к переезду (Сочи, Москва, Санкт-Петербург) , готов к командировкам

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    Список городов-миллионников:
    
   <code>million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    </code>
    
    Инфорация о метро, рядом с которым проживает соискатель нас не интересует.
* Признак **"Готовность к переезду"** должен иметь два возможных варианта: True или False. Обратим внимание, что возможны несколько вариантов описания готовности к переезду в признаке "Город, переезд, командировки". Например:
    * … , готов к переезду , …
    * … , не готова к переезду , …
    * … , готова к переезду (Москва, Санкт-Петербург, Ростов-на-Дону)
    * … , хочу переехать (США) , …
    
    Нас интересует только сам факт возможности или желания переезда.
* Признак **"Готовность к командировкам"** должен иметь два возможных варианта: True или False. Обратим внимание, что возможны несколько вариантов описания готовности к командировкам в признаке "Город, переезд, командировки". Например:
    * … , готов к командировкам , … 
    * … , готова к редким командировкам , …
    * … , не готов к командировкам , …
    
    Нас интересует только сам факт готовности к командировке.
    
    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. По умолчанию будем считать, что такие соискатели не готовы к командировкам.

```python
# создаем функцию
def get_city(arg):
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    city = arg.split(' , ')[0]
    if city in million_cities:
        city = 'город-миллионник'
    elif city == 'Москва' or city == 'Санкт-Петербург':
        return city
    else:
        city = 'другие'
    return city

# создаем новый признак с применением функции
data['Город'] = data['Город, переезд, командировки'].apply(get_city)

print(f"Доля соискателей из Санкт-Петербурга: {round(data['Город'].value_counts(normalize=True)['Санкт-Петербург'] * 100)}%")
```

```python
# создаем новый признак с применением lambda-функции
data['Готовность к переезду'] = data['Город, переезд, командировки'].apply(lambda x: False if 'не готова к переезду' in x or 'не готов к переезду' in x else True)

# создаем функцию
def get_ready_for_trips(arg):
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or('не готова к командировкам' in arg):
            return False
        else:            
            return True
    else:
        return False

# создаем новый признак с применением функции get_ready_for_trips
data['Готовность к командировкам'] = data['Город, переезд, командировки'].apply(get_ready_for_trips)

print(f"Доля соискателей, готовых к переездам и командировкам: {round(data[data['Готовность к переезду'] & data['Готовность к командировкам']].shape[0] / data.shape[0] * 100)}%")
```

5. Признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).

На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу.

Создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

```python
# создаем списки
employment = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedule = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']

# создаем новые признак-мигалки с применением lambda-функции
for emp, sched in zip(employment, schedule):
    data[emp] = data['Занятость'].apply(lambda x: emp in x)
    data[sched] = data['График'].apply(lambda x: sched in x)
    
print(f"Кол-во людей, ищущих проектную работу и волонтерство: {data[data['проектная работа'] & data['волонтерство']].shape[0]}")
print(f"Кол-во людей, желающих работать вахтовым методом и по гибкому графику: {data[data['вахтовый метод'] & data['гибкий график']].shape[0]}")
```

6. Признак заработной платы **"ЗП"**. 
Проблема в том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:
    * 30000 руб.
    * 50000 грн.
    * 550 USD

Сначала приведем заработную плату к единой валюте - рублю. 

Мы уже сделали выгрузку курсов валют, которые встречаются в наших данных за период с 29.12.2017г. по 05.12.2019г.

Итак, возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты. Значение close для рубля заполнить единицей (курс рубля самого к себе).
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратить внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".

```python
# создаем датафрейм с курсами валют
exchange_rate = pd.read_csv('ExchangeRates.csv')
```

```python
exchange_rate.head()
```

```python
# переводим признаки времени в обеих таблицах в формат datetime
data['Обновление резюме'] = pd.to_datetime(data['Обновление резюме'], dayfirst=True).dt.date
exchange_rate['date'] = pd.to_datetime(exchange_rate['date'], dayfirst=True).dt.date

# cоздаем новые признаки для разделения столбца "ЗП"
data['ЗП сумма'] = data['ЗП'].apply(lambda x: x.split()[0]).astype(float)
data['ЗП валюта'] = data['ЗП'].apply(lambda x: x.split()[1])

# приводим значение валюты к стандарту ISO
def get_currency(arg):
    if arg == 'бел.руб.':
        return 'BYN'
    elif arg == 'грн.':
        return 'UAH'
    elif arg == 'сум':
        return 'UZS'
    elif arg == 'руб.':
        return 'RUB'
    else:
        return arg
    return

data['ЗП валюта'] = data['ЗП валюта'].apply(get_currency)

# присоединяем таблицу с курсами валют
merged_data = data.merge(
    exchange_rate,
    left_on=['ЗП валюта', 'Обновление резюме'],
    right_on=['currency', 'date',],
    how='left'
    )
merged_data['close'] = merged_data['close'].fillna(1)
merged_data['proportion'] = merged_data['proportion'].fillna(1)

# высчитываем сумму заработной платы в рублях
data['ЗП (руб)'] = merged_data['ЗП сумма'] * merged_data['close'] / merged_data['proportion']

print(f"Медианная заработная плата: {round(data['ЗП (руб)'].median() / 1000)} тыс. руб.")
```

```python
# удаляем все ненужные признаки
data = data.drop(['Образование и ВУЗ', 'Пол, возраст', 'Опыт работы', 'Город, переезд, командировки', 'Занятость', 'График', 'ЗП', 'ЗП сумма', 'ЗП валюта'], axis=1)
```

3. Итак, у нас есть пропуски в 3-х столбцах: **"Опыт работы (месяц)"**, **"Последнее/нынешнее место работы"**, **"Последняя/нынешняя должность"**. Поступим следующим образом: удалим строки, где есть пропуск в столбцах с местом работы и должностью. Пропуски в столбце с опытом работы заполним **медианным** значением.

```python
# удаляем строки с пропусками в столбцах с местом работы и должностью
data = data.dropna (subset=['Последнее/нынешнее место работы', 'Последняя/нынешняя должность'])

# заменяем пропуски в признаке "Опыт работы (месяц)" на медианное значение
data['Опыт работы (месяц)'] = data['Опыт работы (месяц)'].fillna(data['Опыт работы (месяц)'].median())
print(round(data['Опыт работы (месяц)'].mean()))
```

4. Ликвидация выбросов. Сначала очистим данные вручную. Удалим резюме, в которых указана заработная плата либо выше 1 млн. рублей, либо ниже 1 тыс. рублей.

```python
outliers = data[(data['ЗП (руб)'] > 1e6) | (data['ЗП (руб)'] < 1e3)]
data = data.drop(outliers.index)
print(outliers.shape[0])
```

6. В результате анализа мы обнаружили потенциальные выбросы в признаке **"Возраст"**. Это оказались резюме людей чересчур преклонного возраста для поиска работы. Построим распределение признака в **логарифмическом масштабе**. Добавим к графику линии, отображающие **среднее и границы интервала метода трех сигм**.

Найдем выбросы с помощью метода z-отклонения и удалим их из данных, используем логарифмический масштаб. Сделаем послабление на **1 сигму** (возьмем 4 сигмы) в **правую сторону**.

```python
# строим распределение
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
log_age = np.log(data['Возраст'] + 1)
histplot = sns.histplot(log_age, bins=30, ax=ax)
histplot.axvline(log_age.mean(), color='k', lw=2)
histplot.axvline(log_age.mean() + 4 *log_age.std(), color='k', ls='--', lw=2)
histplot.axvline(log_age.mean() - 3 *log_age.std(), color='k', ls='--', lw=2)
histplot.set_title('Распределение возраста соискателей в логарифмическом масштабе');
```

```python
# вычисляем ассиметрию 
if log_age.skew() > 0:
    print('Асимметрия логарифмического распределения правосторонняя')
else:
    print('Асимметрия логарифмического распределения левосторонняя')
```

```python
# находим количество выбросов по методу z-отклонений
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score(data, 'Возраст', log_scale=True)
print(outliers.shape[0])
```
